```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests

def get_morningstar_financials():
    stock = input('請輸入要查詢美股名稱：')
    url = 'http://financials.morningstar.com/finan/financials/getFinancePart.html?&t=XNYS:'+stock
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text, 'lxml')
    Table = soup.find('table')
    Table = str(Table)
    Table = Table.replace('\'', '')
    Table = Table.replace('\\', '')
    soup = BeautifulSoup(Table,'lxml')
    aaa = soup.find('table', {'class':'r_table1'})

    contents = []
    for ele in aaa.findAll('tr'):
        contents.append(ele)

    ccss = []
    for ele in aaa.findAll('th'):
        ccss.append(ele.text)
    list_1 = ccss[1:12]
    list_2 = ccss[12:] 

    Revenue = ['Revenue']    
    for i in contents[2].findAll('td'):
        Revenue.append(i.text)

    GrossMargin = ['Gross Margin %']
    for i in contents[4].findAll('td'):
        GrossMargin.append(i.text)

    OperatingIncome = ['Operating Income']
    for i in contents[6].findAll('td'):
        OperatingIncome.append(i.text)

    OperatingMargin = ['Operating Margin %']
    for i in contents[8].findAll('td'):
        OperatingMargin.append(i.text)

    NetIncome = ['Net Income'] 
    for i in contents[10].findAll('td'):
        NetIncome.append(i.text)

    EarningsPerShare = ['Earnings Per Share']    
    for i in contents[12].findAll('td'):
        EarningsPerShare.append(i.text)

    Dividends = ['Dividends']
    for i in contents[14].findAll('td'):
        Dividends.append(i.text)

    PayoutRatio = ['Payout Ratio %'] 
    for i in contents[16].findAll('td'):
        PayoutRatio.append(i.text)

    Shares = ['Shares']
    for i in contents[18].findAll('td'):
        Shares.append(i.text)

    BookValuePerShare = ['Book Value Per Share']
    for i in contents[20].findAll('td'):
        BookValuePerShare.append(i.text)

    OperatingCashFlow = ['Operating Cash Flow']
    for i in contents[22].findAll('td'):
        OperatingCashFlow.append(i.text)

    CapSpending = ['Cap Spending']
    for i in contents[24].findAll('td'):
        CapSpending.append(i.text)

    FreeCashFlow = ['Free Cash Flow']
    for i in contents[26].findAll('td'):
        FreeCashFlow.append(i.text)

    FreeCashFlowPerShare = ['Free Cash Flow Per Share']
    for i in contents[28].findAll('td'):
        FreeCashFlowPerShare.append(i.text)

    WorkingCapital = ['Working Capital']
    for i in contents[30].findAll('td'):
        WorkingCapital.append(i.text)

    df = pd.DataFrame([Revenue, GrossMargin, OperatingIncome, OperatingMargin, NetIncome, EarningsPerShare, 
                       Dividends, PayoutRatio, Shares, BookValuePerShare, OperatingCashFlow, CapSpending,
                       FreeCashFlow, FreeCashFlowPerShare, WorkingCapital], index = list_2)

    df.drop([0],axis = 1, inplace = True) 
    df.columns = list_1
    return df
```

```python
get_morningstar_financials()
```
